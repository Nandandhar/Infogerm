```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\NANDAN's Documents\Infogerm int\Coffee sales.csv")
data
```

```python
coffee_brown = '#6F4E37'
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])

daily_sales = data.groupby(data['datetime'].dt.date)['money'].sum()

plt.figure(figsize=(10, 6))
plt.plot(daily_sales.index, daily_sales.values, marker='o', linestyle='-', color=coffee_brown)
plt.title('Daily Coffee Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales (Money)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
coffee_popularity = data['coffee_name'].value_counts()

plt.figure(figsize=(10, 6))
coffee_popularity.plot(kind='bar', color=coffee_brown)
plt.title('Most Popular Coffee Types')
plt.xlabel('Coffee Type')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sales_by_coffee_type = data.groupby('coffee_name')['money'].sum()

plt.figure(figsize=(10, 6))
sales_by_coffee_type.plot(kind='bar', color=coffee_brown)
plt.title('Total Sales by Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Total Sales (Money)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
hourly_sales = data.groupby(data['datetime'].dt.hour)['money'].sum()

plt.figure(figsize=(10, 6))
hourly_sales.plot(kind='line', marker='o', linestyle='-', color=coffee_brown)
plt.title('Hourly Coffee Sales')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales (Money)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
payment_distribution = data['cash_type'].value_counts()

plt.figure(figsize=(10, 6))
payment_distribution.plot(kind='bar', color=coffee_brown)
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```
